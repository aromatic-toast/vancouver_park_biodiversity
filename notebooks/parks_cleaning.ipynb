# Parks Data Cleaning 

## import libraries 

```python
import pandas as pd
import geopandas as gpd
from mealprep.mealprep import find_missing_ingredients
```

```python
parks = gpd.read_file('../data/raw_data/vancouver_parks.shp')
```

```python
parks.head()
```

```python
# remove columns 
parks = parks[['PARK_NAME', 'PARK_TYPE', 'PARK_PRIMA', 'LATITUDE', 'LONGITUDE', 'geometry']]
```

```python
parks.head()
```

```python
# rename columns 
parks = parks.rename(columns={'PARK_NAME':'park_name', 'PARK_TYPE': 'park_type', 'PARK_PRIMA':'park_primary_use',
                      'LATITUDE': 'latitude', 'LONGITUDE':'longitude'})
```

```python
parks.head()
```

```python
find_missing_ingredients(parks)
```

```python
# save clean parks file 
parks.to_file('../data/clean_data/vancouver_parks.shp')
```
